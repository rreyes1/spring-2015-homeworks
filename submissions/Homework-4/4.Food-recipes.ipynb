```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
def addItemToDict( my_map, k, v):
    if k in my_map:
        l = my_map[k]
        if v in l:
            u = l.get(v) + 1
            my_map[k][v] = u
        else:
            my_map[k][v] =1
    else: 
        my_map[k] = {v: 1}
```

```python
menu = pd.read_csv('menu_recipes.txt', names='r')
menu_map = dict()
for idx, row in menu.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    #print nation
    recipe = string[i+1:]
    i = recipe.find('\t')
    if i != -1:
        item = recipe[0:i]
    else:
        item = recipe[0:]
    addItemToDict(menu_map, nation, item)
    i = 0
    while recipe.find('\t',i+1) != -1:
        i = recipe.find('\t',i+1)
        j = recipe.find('\t',i+1)
        if j != -1:
            item = recipe[i+1:j]
        else:
            item = recipe[i+1:]
        addItemToDict(menu_map, nation, item)
#print menu_map
```

```python
epic = pd.read_csv('epic_recipes.txt', names='r')
epic_map = dict()
for idx, row in epic.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    #print nation
    recipe = string[i+1:]
    i = recipe.find('\t')
    if i != -1:
        item = recipe[0:i]
    else:
        item = recipe[0:]
    addItemToDict(epic_map, nation, item)
    i = 0
    while recipe.find('\t',i+1) != -1:
        i = recipe.find('\t',i+1)
        j = recipe.find('\t',i+1)
        if j != -1:
            item = recipe[i+1:j]
        else:
            item = recipe[i+1:]
        addItemToDict(epic_map, nation, item)
```

```python
allr = pd.read_csv('allr_recipes.txt', names='r')
allr_map = dict()
for idx, row in allr.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    #print nation
    recipe = string[i+1:]
    i = recipe.find('\t')
    if i != -1:
        item = recipe[0:i]
    else:
        item = recipe[0:]
    addItemToDict(allr_map, nation, item)
    i = 0
    while recipe.find('\t',i+1) != -1:
        i = recipe.find('\t',i+1)
        j = recipe.find('\t',i+1)
        if j != -1:
            item = recipe[i+1:j]
        else:
            item = recipe[i+1:]
        addItemToDict(allr_map, nation, item)
```

```python
regions = pd.read_csv('map.txt', names='r')
regions_map = dict()
for idx, row in regions.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    region = string[i+1:]
    regions_map[nation] = region

#for i in regions_map:
#    print i, regions_map[i]
```

```python
menu = pd.DataFrame(menu_map)
allr = pd.DataFrame(allr_map)
epic = pd.DataFrame(epic_map)
menu = menu.fillna(0)
allr = allr.fillna(0)
epic = epic.fillna(0)
recipes = menu.add(allr,fill_value=0).add(epic,fill_value=0)
recipes = recipes.fillna(0)
recipes.head()
```
