```python
import time

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as datasets
import sklearn.metrics as metrics
import scipy.sparse.linalg as linalg
import scipy.cluster.hierarchy as hr

import sklearn.cluster as cluster

from sklearn.decomposition import TruncatedSVD

from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler

#import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
def addItemToDict( my_map, k, v):
    if k in my_map:
        l = my_map[k]
        if v in l:
            u = l.get(v) + 1
            my_map[k][v] = u
        else:
            my_map[k][v] =1
    else: 
        my_map[k] = {v: 1}
```

```python
menu = pd.read_csv('menu_recipes.txt', names='r')
menu_map = dict()
for idx, row in menu.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    #print nation
    recipe = string[i+1:]
    i = recipe.find('\t')
    if i != -1:
        item = recipe[0:i]
    else:
        item = recipe[0:]
    addItemToDict(menu_map, nation, item)
    i = 0
    while recipe.find('\t',i+1) != -1:
        i = recipe.find('\t',i+1)
        j = recipe.find('\t',i+1)
        if j != -1:
            item = recipe[i+1:j]
        else:
            item = recipe[i+1:]
        addItemToDict(menu_map, nation, item)
#print menu_map
```

```python
epic = pd.read_csv('epic_recipes.txt', names='r')
epic_map = dict()
for idx, row in epic.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    #print nation
    recipe = string[i+1:]
    i = recipe.find('\t')
    if i != -1:
        item = recipe[0:i]
    else:
        item = recipe[0:]
    addItemToDict(epic_map, nation, item)
    i = 0
    while recipe.find('\t',i+1) != -1:
        i = recipe.find('\t',i+1)
        j = recipe.find('\t',i+1)
        if j != -1:
            item = recipe[i+1:j]
        else:
            item = recipe[i+1:]
        addItemToDict(epic_map, nation, item)
```

```python
allr = pd.read_csv('allr_recipes.txt', names='r')
allr_map = dict()
for idx, row in allr.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    #print nation
    recipe = string[i+1:]
    i = recipe.find('\t')
    if i != -1:
        item = recipe[0:i]
    else:
        item = recipe[0:]
    addItemToDict(allr_map, nation, item)
    i = 0
    while recipe.find('\t',i+1) != -1:
        i = recipe.find('\t',i+1)
        j = recipe.find('\t',i+1)
        if j != -1:
            item = recipe[i+1:j]
        else:
            item = recipe[i+1:]
        addItemToDict(allr_map, nation, item)
```

```python
regions = pd.read_csv('map.txt', names='r')
regions_map = dict()
for idx, row in regions.iterrows():
    string = row['r']
    i = string.find('\t')
    nation = string[0:i]
    region = string[i+1:]
    regions_map[nation] = region
regions_map['Asian'] = "Asian"
regions_map['asian'] = "Asian"

#for i in regions_map:
#    print i, regions_map[i]
```

```python
menu = pd.DataFrame(menu_map)
allr = pd.DataFrame(allr_map)
epic = pd.DataFrame(epic_map)
menu = menu.fillna(0)
allr = allr.fillna(0)
epic = epic.fillna(0)
recipes = menu.add(allr,fill_value=0).add(epic,fill_value=0)
recipes = recipes.fillna(0)
recipes.head()
```

```python
# Create a Table that has nations as rows and ingredients as features:
df = recipes.T
df.head()
```

```python
X = StandardScaler().fit_transform(df)
sns.heatmap(X,xticklabels=False, yticklabels=False, linewidths=0)
X
```

```python
distanceMatrix = pdist(X,'euclidean')
Z = hr.linkage(X, method='complete', metric='euclidean')
print Z.shape, X.shape
fig = plt.figure(figsize=(10,10))
T = hr.dendrogram(Z,color_threshold=0.4, leaf_font_size=4)
fig.show()
```

```python
distances = metrics.euclidean_distances(X)
cg = sns.clustermap(distances, method="complete", figsize=(13,13), xticklabels=False)
print cg.dendrogram_col.reordered_ind
```

```python
# We should cluster by regions potentially:
regionsdf = pd.DataFrame()
for i in recipes:
    c = regions_map.get(i)
    if c in regionsdf:
        regionsdf[c].add(recipes[i])
    else:
        regionsdf[c] = recipes[i]

regionsdf = regionsdf.T
regionsdf
```

```python
X = StandardScaler().fit_transform(regionsdf)
distanceMatrix = pdist(X,'euclidean')
Z = hr.linkage(X, method='complete', metric='euclidean')
print Z.shape, X.shape
fig = plt.figure(figsize=(10,10))
T = hr.dendrogram(Z,color_threshold=0.4, leaf_font_size=4)
fig.show()
distances = metrics.euclidean_distances(X)
cg = sns.clustermap(distances, method="complete", figsize=(13,13), xticklabels=False)
print cg.dendrogram_col.reordered_ind

```
